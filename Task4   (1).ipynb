```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('USvideos.csv')

# Display basic information about the dataset
print("Dataset Overview:")
print(data.info())

# Check column names and ensure numeric conversion where needed
numeric_columns = ['views', 'likes', 'dislikes', 'comment_count']
for col in numeric_columns:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows with missing values in numeric columns
data = data.dropna(subset=numeric_columns)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(data['views'], bins=30, color='blue', alpha=0.7, label='Views')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(data['likes'], bins=30, color='green', alpha=0.7, label='Likes')
plt.title('Distribution of Likes', fontsize=16)
plt.xlabel('Likes', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(data['dislikes'], bins=30, color='red', alpha=0.7, label='Dislikes')
plt.title('Distribution of Dislikes', fontsize=16)
plt.xlabel('Dislikes')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
