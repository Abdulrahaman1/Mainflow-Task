```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('householdtask3.csv')
```

```python
data.head(15)
```

```python
# Bar Chart: Total households across years
plt.figure(figsize=(10, 6))
plt.bar(data['year'], data['own'], color='skyblue', edgecolor='black')
plt.title('Bar Chart')
plt.xlabel('Year')
plt.ylabel('Own')
plt.xticks(data['year'])
plt.tight_layout()
plt.show()
```

```python
# Line Chart Year against own
plt.figure
plt.plot(data['year'], data['own'], color='green')
plt.title('Line chart', fontsize=16)
plt.xlabel('Year')
plt.ylabel('own')
plt.tight_layout()
plt.show()
```
